# For Loops 

```python
array = np.array([3,4,6,7,8,9])
```

```python
for i in array:
    print('value is ',i)
```

```python
arr3 = np.array([[200,300],[400,500]])
print('Array1 is :\n',arr3)

for item in arr3:
    print('value is ',item)
```

```python
arr3 = np.array([[200,300],[400,500]])
print('Array1 is :\n',arr3)

for i in arr3:
    for j in i:
        print('value is ',j)
```

```python
arr3 = np.array([[[3,4],[5,6]]])
print('Array1 is :\n',arr3)

for i in arr3:
    for j in i:
        for k in j:
            print('value is ',k)
```

# nditer()

```python
arr3 = np.array([[200,300],[400,500]])
print('Array1 is :\n',arr3)

for item in np.nditer(arr3):
    print('value is ',item)
```

```python
arr3 = np.array([[[3,4],[5,6]]])
print('Array1 is :\n',arr3)

for item in np.nditer(arr3):
    print('value is ',item)
```

```python
arr3 = np.array([[200,300],[400,500]])
print('Array1 is :\n',arr3)

for i,item in np.ndenumerate(arr3):
    print(f'value at index {i} is {item}')
```

```python
arr3 = np.array([[[3,4],[5,6]],[[30,40],[50,60]]])
print('Array1 is :\n',arr3)

for i,item in np.ndenumerate(arr3):
    print(f'value at index {i} is {item}')
```

# np.where()

```python
l1 = [2,3,4,5,4,6,4,8,9]
l1.index(4)

indices = [i for i,item in enumerate(l1) if item == 4]
indices
```

```python
arr1 = np.array([2,3,4,5,4,6,4,8,9])

# indices = np.where(condition)
indices = np.where(arr1 == 4)
indices
```

```python
arr1 = np.array([2,3,4,5,4,6,4,8,9])

# indices = np.where(condition)
indices = np.where(arr1 <= 4)
indices
```
